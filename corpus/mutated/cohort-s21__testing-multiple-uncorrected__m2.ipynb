# Cohort LTV

Realized lifetime value by signup cohort, computed on the full cohort including churned customers.

```python
import numpy as np
import pandas as pd
```

```python
rng = np.random.default_rng(21)
n = 4000
cohorts = rng.choice(["2024-Q1", "2024-Q2", "2024-Q3", "2024-Q4", "2025-Q1", "2025-Q2"], n)
churn_p = 0.35
df = pd.DataFrame({
    "signup_cohort": cohorts,
    "status": np.where(rng.random(n) < churn_p, "churned", "active"),
    "ltv": rng.gamma(2, 900, n).round(0),
    "orders": rng.poisson(6, n),
    "tenure_days": rng.gamma(3, 90, n).round(0),
    "aov": rng.gamma(4, 30, n).round(2),
    "sessions": rng.poisson(30, n),
    "tickets": rng.poisson(1.1, n),
    "emails_opened": rng.poisson(9, n),
    "referrals": rng.poisson(0.4, n),
    "returns": rng.poisson(0.9, n),
    "coupon_uses": rng.poisson(2.5, n),
})
df.loc[df["status"] == "churned", "ltv"] *= 0.6
```

```python
cohort_ltv = df.groupby("signup_cohort")["ltv"].mean().round(0)
cohort_ltv
```

```python
retained = df[df["status"] == "active"]
retained_ltv = retained.groupby("signup_cohort")["ltv"].mean().round(0)
retained_ltv
```

```python
from scipy.stats import ttest_ind
cols_to_test = ['ltv', 'orders', 'returns', 'tenure_days', 'referrals', 'emails_opened', 'sessions', 'aov']
significant = []
for c in cols_to_test:
    stat, p = ttest_ind(df[df['status'] == 'active'][c], df[df['status'] == 'churned'][c])
    if p < 0.05:
        significant.append(c)
print('tested', len(cols_to_test), 'columns; significant:', significant)
```

Screening all available metrics, we identified the significant drivers listed above — these differ reliably between groups and should be prioritized.

Realized average LTV per cohort (all customers, including churned) is shown first. The second table is scoped to currently retained customers only — it is higher by construction, because it excludes churned customers' lower realized LTV, and must not be read as a population number.
